```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
incandescent_data = pd.read_csv("Incandescent.CSV")
incandescent_t = np.array(incandescent_data['t'])
incandescent_v = np.array(incandescent_data['v'])

incandescent_v_scale = 5.000000e-01
incandescent_t_scale = 5.000000e-03

incan_x1, incan_y1 = (-00.022300000000,   0.52000)
incan_x2, incan_y2 = (-00.014160000000,   0.54000)

incan_t = incan_x2 - incan_x1
incan_f = 1/incan_t
print(incan_f)

plt.figure(figsize=(16,12))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title(r"Measured voltage of a photodiode under an incandescent light bulb measured by an oscilloscope.")


plt.xlabel(r"Time (s)", fontsize=18)
plt.ylabel(r"Voltage (V)", fontsize=18)

plt.scatter(incandescent_t, incandescent_v, label="Data Point")
plt.legend(loc="best", prop={"size": 20})
plt.show()
```

```python
cfl_high_data = pd.read_csv("CFL_High.CSV")
cfl_low_data = pd.read_csv("CFL_Low.CSV")

cfl_high_t = np.array(cfl_high_data['t'])
cfl_high_v = np.array(cfl_high_data['v'])

high_x1, high_y1 = (0.000055580000,  -0.05600)
high_x2, high_y2 = (0.000066460000,  -0.04000)

high_t = high_x2 - high_x1
high_f = 1 / high_t
print(high_f)


plt.figure(figsize=(16,12))
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title(r"High frequency voltage of a photodiode under a compact flourescent bulb measured by an oscilloscope.")

plt.xlabel(r"Time (s)", fontsize=18)
plt.ylabel(r"Voltage (V)", fontsize=18)

plt.scatter(cfl_high_t, cfl_high_v, label="Data Point")
plt.legend(loc="best", prop={"size": 20})
plt.show()
```
